```python
import pandas as pd 
```

### Carga de Datos 
Carga de archivo <br>
Renombrar columnas <br>
Seleccionar columnas necesarias para el analisis<br>


```python
df=pd.read_excel("Data\\Pobreza.xlsx", sheet_name=2,skiprows=8)
df=df.rename(columns={"Unnamed: 0":"Año","Unnamed: 1":"Indicador"})
df=df[["Año","Indicador","Azuay", "Bolívar", "Cañar", "Carchi", "Cotopaxi", "Chimborazo", "El Oro", "Esmeraldas", "Guayas", "Imbabura", 
            "Loja", "Los Ríos", "Manabí", "Morona Santiago", "Napo", "Pastaza", "Pichincha", "Tungurahua", "Zamora Chinchipe", 
            "Sucumbíos", "Orellana", "Santo Domingo", "Santa Elena"]]

```

### Manipulacion del dataframe

Normalizar datos : Las provincias se encuentran en columnas y no en filas . 

```python
df_unpivot=df.melt(id_vars=["Año","Indicador"],
                   var_name="Provincia",
                   value_name="Valor")

df_unpivot
```

Pivotear dataframe para colocar como columnas los indicadores de pobreza 

```python
df_pivot=df_unpivot.pivot_table(index=["Año","Provincia"],
                                columns="Indicador",
                                values="Valor").reset_index()

df_pivot
```

```python
Provincia_Pobreza=df_pivot
Provincia_Pobreza["Provincia"]=Provincia_Pobreza["Provincia"].str.upper()
Provincia_Pobreza
```
