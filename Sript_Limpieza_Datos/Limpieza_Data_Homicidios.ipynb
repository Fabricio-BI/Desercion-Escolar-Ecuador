```python
import pandas as pd

```

## Carga de los Datos

```python
df=pd.read_excel("Data\\Homicidios_2014_2024.xlsx")
```

```python
## Agregamos una columna con el año 
df["Año"]=df["fecha_infraccion"].dt.year 

df
```

### Crear un dataframe que agrupé los datos por Provincia 

```python
df_provinciales=df.groupby(["Año","provincia"]).size().reset_index(name="Total_Homicidios")
df_provinciales.columns=df_provinciales.columns.str.replace(" "," ")
df_provinciales=df_provinciales.rename(columns={
    "provincia":"Provincia"
})
df_provinciales
```

```python
df_provinciales

reemplazos={"BOLIVAR":"BOLÍVAR",
            "CANAR":"CAÑAR",
            "LOS RIOS":"LOS RÍOS",
            "MANABI":"MANABÍ",
            "SUCUMBIOS":"SUCUMBÍOS"
            }

df_provinciales["Provincia"]=df_provinciales["Provincia"].replace(reemplazos,regex=False)
df_provinciales
```

### Se importa los datos poblacionales de cada provincia para calcular la tasa de homicidio por 100.mil habitantes 

```python
pb=pd.read_excel("Data\\Poblacion.xlsx",sheet_name=3,skiprows=1)

pb=pb.rename(columns={
    "Unnamed: 0":"Provincia"
})

pb["Total_Habitantes"]=pb["Hombre"]+pb["Mujer"]+pb["Indefinido1"]
pb["Provincia"]=pb["Provincia"].str.upper()

pb["Provincia"]=pb["Provincia"].str.replace("SANTO DOMINGO DE LOS TSÁCHILAS","STO DGO DE LOS TSACHILAS",regex=False)
pb["Provincia"]=pb["Provincia"].str.replace("NO REGISTRA2","ZONA NO DELIMITADA",regex=False)
pb
```

```python
from pandasgui import show
```

```python
Tasa_provincial=pd.merge(df_provinciales,pb ,on="Provincia",how="left")
Tasa_provincial["Total_Habitantes"]=Tasa_provincial["Total_Habitantes"].fillna(0).astype(int)
Tasa_provincial
```

### Crear funcion para calcular tasa de homicidios 

```python
def tasa_homicidios(homicidios,habitantes):
    if habitantes==0 or pd.isnull(habitantes):
        return 0
    tasa=(homicidios/habitantes)*100000
    tasa=round(tasa,1)
    return tasa 
```

### Aplicar funcion al dataframe

```python
Tasa_provincial["Tasa_Homicidios"]= Tasa_provincial.apply(
    lambda row:tasa_homicidios(row["Total_Homicidios"],row["Total_Habitantes"]),axis=1
)

Tasa_provincial=Tasa_provincial[["Año","Provincia","Total_Homicidios","Total_Habitantes","Tasa_Homicidios"]]

Tasa_provincial
```

```python
Tasa_provincial.to_csv("Data_Procesada\\Tasa_Homicidios.csv",index=False)
```
