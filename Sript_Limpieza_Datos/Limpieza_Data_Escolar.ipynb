#### Limpieza y manipulacion de los datos correspondientes al periodo lectivo 2023-2024

Convertir el dataframe de formato ancho , en formato largo 

```python
Unpivot_2023=df_2023.melt(id_vars=[
    "Año_Lectivo", "Zona  ", "Provincia", "Cod_Provincia", "Cantón", "Cod_Cantón", 
    "Parroquia", "Cod_Parroquia", "Nombre Institución", "Tipo Educación", 
    "Nivel Educación", "Sostenimiento", "Área", "Régimen Escolar", "Jurisdicción", 
    "Modalidad", "Jornada", "Tenencia del inmueble", "Forma Acceso", 
    "Docentes_Femenino", "Docentes_Masculino", "Total_Docentes", 
    "Administrativos_Femenino", "Administrativos_Masculino", 
    "Total_Administrativos", "Estudiantes_Femenino", "Estudiantes_Masculino", 
    "Total_Estudiantes"
],
                          var_name="Atributo",
                          value_name="Valor")
```


Extraer categorias para luego volver pivotar la tabla y totalizar los alunnos por categoria 

```python
Unpivot_2023["Categoria"]=Unpivot_2023["Atributo"].str.extract(r'(Promovidos|NoPromovidos|Abandono)')
Unpivot_2023["Categoria"]=Unpivot_2023["Categoria"].fillna("Total")
Unpivot_2023["Valor"]=Unpivot_2023["Valor"].replace("-",0)
Unpivot_2023["Categoria"].unique()
```

```python
Pivot_2023=Unpivot_2023.pivot_table(index=["Año_Lectivo", "Zona  ", "Provincia", "Cod_Provincia", "Cantón", "Cod_Cantón", 
                                            "Parroquia", "Cod_Parroquia", "Nombre Institución", "Tipo Educación", 
                                            "Nivel Educación", "Sostenimiento", "Área", "Régimen Escolar", "Jurisdicción", 
                                            "Modalidad", "Jornada", "Tenencia del inmueble", "Forma Acceso", 
                                            "Docentes_Femenino", "Docentes_Masculino", "Total_Docentes", 
                                            "Administrativos_Femenino", "Administrativos_Masculino", 
                                            "Total_Administrativos", "Estudiantes_Femenino", "Estudiantes_Masculino", 
                                            "Total_Estudiantes"],
                                    columns="Categoria",
                                    values="Valor",
                                    aggfunc="sum").reset_index()
```

Limpiar columnas Año Lectivo  y añadir columna Año calendario 

```python
Pivot_2023["Año"]=Pivot_2023["Año_Lectivo"].str.extract(r'(\d{4})')
Pivot_2023["Año_Lectivo"]=Pivot_2023["Año_Lectivo"].str.replace("Fin","")
Pivot_2023["Año_Lectivo"]=Pivot_2023["Año_Lectivo"].str.replace("Fin","")
Pivot_2023.head(5)
```

Escoger columnas necesarias para el analisis

```python
Edu_2023=Pivot_2023[["Año","Año_Lectivo","Zona  ","Provincia","Cantón","Sostenimiento", "Área", "Régimen Escolar", "Jurisdicción","Total","Promovidos","NoPromovidos","Abandono"]]
Edu_2023.head(2)
```

#### Crear nuevo dataframe anexando los datos del 2023 a la serie historica del 2009-2022
